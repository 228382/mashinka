Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
for i in range(12):
    df[f'k{i}'] = df['Задания с кратким ответом'].apply(lambda x: 1 if x[i] == '+' else 0)
```

```python
for i in range(12):
    print(f'{i+1}) Процент выполнения : ',np.round(100*df[f'k{i}'].sum()/df.shape[0],1))
```

```python
for i in range(12):
    print(f'{i+1}) Процент не выполнения : ',np.round(100*(1-df[f'k{i}']).sum()/df.shape[0],1))
```

Аналогично и с типом С (ответы с развернутым ответом)

```python
for i in range(5):
    df[f'k_1{i}'] = df['Задания с развёрнутым ответом'].apply(lambda x: 1 if x[4*i] != '0' else 0)
```

```python
for i in range(5):
    print(f'{i+1}) Процент выполнения: ',np.round(100*df[f'k_1{i}'].sum()/df.shape[0],1))
```

```python
for i in range(5):
    print(f'{i+1}) Процент не выполнения: ',np.round(100*(1-df[f'k_1{i}']).sum()/df.shape[0],1))
```
